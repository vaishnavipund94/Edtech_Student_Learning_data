```python
#34. Create a new column total_score as the sum of quiz, assignment, and project scores.
df["total_score"] = df["avg_quiz_score"] + df["assignment_score"] + df["project_score"]
```

```python
#Q35. Create a new column attendance_flag based on attendance percentage.
def attendance_label(x):
    if x >= 75:
        return "Good"
    else:
        return "Poor"

df["attendance_flag"] = df["attendance_pct"].apply(attendance_label)
```

```python
df.head()
```

```python
#Q36. Create a new column learning_intensity based on learning hours.
def learning_hrs(x):
    if x > 20:
        return "High"
    elif x >= 10:
        return "Medium"
    else:
        return "Low"

df["learning_intensity"] = df["learning_hours"].apply(learning_hrs)
```

```python
df.head()
```

```python
#37. Classify students based on total_score using apply().

def student_performance(x):
    if x >= 250:
        return "Excellent"
    elif x >= 200:
        return "Good"
    else:
        return "Poor"

df["performance"] = df["total_score"].apply(student_performance)
```

```python
df.head()
```

```python
#38. Create a placement eligibility column using apply().

def eligible(x):
    if x["attendance_pct"] >= 75 and x["total_score"] >= 200:
        return "Eligible"
    else:
        return "Not Eligible"

df["placement_eligibilty"] = df.apply(eligible, axis=1)
```

```python
df.head()
```

```python
#39. Map placement_status values to numerical values.

placement = {"Placed" : 1, "Not Placed" : 0}

df["placement_numeric_status"] = df["placement_status"].map(placement)
    
```

```python
df.head()
```

```python
#40. Map batch_type values to abbreviated codes.
code = {"weekend" : "WE", "weekday" : "WD"}
df["batch_code"] = df["batch_type"].map(code)
```

```python
df.head()
```

```python
#41. Map gender values to single-letter codes.
gender = {"Male" : "M", "Female":"F"}
df["gender_code"] = df["gender"].map(gender)
```
